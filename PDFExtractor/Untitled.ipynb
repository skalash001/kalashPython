```python
import PyPDF2
from PIL import Image

# Open the PDF file in read-binary mode
with open('example.pdf', 'rb') as pdf_file:
    # Create a PDF reader object
    pdf_reader = PyPDF2.PdfFileReader(pdf_file)

    # Get the number of pages in the PDF file
    num_pages = len(pdf_reader.pages)

    # Prompt the user for a page number to extract
    page_num = int(input(f"Enter a page number to extract (1-{num_pages}): ")) - 1

    # Get the specified page object
    page_obj = pdf_reader.pages[page_num]

    # Get the first image object on the page
    try:
        image_obj = next(obj for obj in page_obj.objects if isinstance(obj, PyPDF2.pdf.ContentStream._XObject) and obj.name == '/Im0')
    except StopIteration:
        print(f"No image found on page {page_num + 1}")
    else:
        # Extract the image data from the page
        image_stream = image_obj.stream.get_rawdata()
        image = Image.open(image_stream)

        # Save the extracted image as a JPEG file
        image_filename = f"page{page_num + 1}_image.jpg"
        image.save(image_filename, "JPEG")
        print(f"Image saved as {image_filename} in the current directory.")
```

```python
import PyPDF2
import os

# Get the filepath of the PDF file to extract from
pdf_filepath = 'C:\\Users\\skala\\Downloads\\Kanvasatapathabrahmaaam_Vol_VI.pdf'

# Open the PDF file in read-binary mode
with open(pdf_filepath, 'rb') as pdf_file:
    # Create a PDF reader object
    pdf_reader = PyPDF2.PdfReader(pdf_file)

    # Get the number of pages in the PDF file
    num_pages = len(pdf_reader.pages)

    # Prompt the user for a page number range to extract
    start_page_num = int(input(f"Enter the starting page number to extract (1-{num_pages}): ")) - 1
    end_page_num = int(input(f"Enter the ending page number to extract (1-{num_pages}): "))

    # Create the output directory if it doesn't already exist
    output_dir = 'G:/PDF_Extractor/Extracted'
    os.makedirs(output_dir, exist_ok=True)

    # Loop through each page in the specified range
    for page_num in range(start_page_num, end_page_num):
        # Get the page object
        page = pdf_reader.pages[page_num]

        # Create a PDF writer object
        pdf_writer = PyPDF2.PdfWriter()

        # Add the page to the PDF writer
        pdf_writer.addpage(page)

        # Save the PDF writer contents to a new file
        output_filepath = os.path.join(output_dir, f"page{page_num + 1}.pdf")
        with open(output_filepath, 'wb') as output_file:
            pdf_writer.write(output_file)

        print(f"Page {page_num + 1} extracted and saved as {output_filepath}")
```

```python
# Import the necessary modules
import os
import PyPDF2

# Prompt the user for input
input_filepath = 'C:\\Users\\skala\\Downloads\\Kanvasatapathabrahmaaam_Vol_VI.pdf'
start_page_num = int(input("Enter the starting page number to extract (1-458): ")) - 1
end_page_num = int(input("Enter the ending page number to extract (1-458): ")) - 1

# Set the output directory
output_dir = "G:/PDF_Extractor/Extracted"

# Create the output directory if it does not exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Open the PDF file in read-binary mode
with open(input_filepath, 'rb') as pdf_file:
    # Create a PDF reader object
    pdf_reader = PyPDF2.PdfFilReader(pdf_file)

    # Get the number of pages in the PDF file
    num_pages = pdf_reader.getNumPages()

    # Ensure that the start page and end page numbers are within the bounds of the PDF
    start_page_num = max(0, min(start_page_num, num_pages - 1))
    end_page_num = max(0, min(end_page_num, num_pages - 1))

    # Loop through the range of pages to extract
    for page_num in range(start_page_num, end_page_num + 1):
        # Get the page object
        page = pdf_reader.getPage(page_num)

        # Create a PDF writer object
        pdf_writer = PyPDF2.PdfFileWriter()

        # Add the page to the PDF writer
        pdf_writer.addPage(page)

        # Save the PDF writer contents to a new file
        output_filepath = os.path.join(output_dir, f"page{page_num + 1}.pdf")
        with open(output_filepath, 'wb') as output_file:
            pdf_writer.write(output_file)

print(f"Pages {start_page_num + 1} to {end_page_num + 1} were successfully extracted to {output_dir}.")
```

```python
import PyPDF2

def remove_alternate_pages(file_path):
    start_page = int(input('Enter the page number to start removing alternate pages from: '))
    with open(file_path, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfFileReader(pdf_file)
        pdf_writer = PyPDF2.PdfFileWriter()
        for page_num in range(pdf_reader.numPages):
            if page_num < start_page or page_num % 2 == 0:
                pdf_writer.addPage(pdf_reader.getPage(page_num))
        new_file_path = file_path.split('.pdf')[0] + '_new.pdf'
        with open(new_file_path, 'wb') as new_file:
            pdf_writer.write(new_file)
    print(f'New file saved at {new_file_path}')

remove_alternate_pages('C:\\Users\\skala\\Downloads\\Kanvasatapathabrahmaaam_Vol_VI.pdf')
```
